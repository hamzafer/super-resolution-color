```python
from pathlib import Path
from PIL import Image

def print_distinct_image_sizes(directory_path):
    # Convert to Path object for easier handling
    directory = Path(directory_path)

    # Check if directory exists
    if not directory.is_dir():
        print(f"The directory {directory_path} does not exist.")
        return

    # Collect distinct sizes
    sizes = set()
    for image_path in directory.glob('*'):
        try:
            with Image.open(image_path) as img:
                sizes.add(img.size)
        except Exception as e:
            print(f"Could not open {image_path.name}: {e}")

    # Print distinct sizes
    if sizes:
        print(f"Distinct image sizes in {directory_path}:")
        for size in sizes:
            print(f"  {size[0]}x{size[1]}")
    else:
        print(f"No images found in {directory_path}.")

# Check distinct image sizes in both directories
print("Checking sizes in '/home/hamzaz/hamza/super-resolution-color/dataset/imagenet256/lq'")
print_distinct_image_sizes("/home/hamzaz/hamza/super-resolution-color/dataset/imagenet256/lq")

print("\nChecking sizes in 'testdata/RealSet80'")
print_distinct_image_sizes("/home/hamzaz/hamza/super-resolution-color/models/ResShift/testdata/RealSet80")
```
